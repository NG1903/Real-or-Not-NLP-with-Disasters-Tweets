```python
#code for shuffling data

import random
import csv
filename= 'train (2).csv'

with open(filename,'r',encoding="utf8") as csvfile:
        readCSV=csv.reader(csvfile)
        tweets=list()
        for row in readCSV:
            tweets.append(row)
        random.shuffle(tweets)
        with open('train_tweets.csv','w',encoding="utf8",newline='') as file:
            writer=csv.writer(file)
            for row in tweets:
                writer.writerow(row)
```
